```python
import pdfplumber #Imports the pdfplumber library which is used to extract text from PDF documents
from transformers import pipeline #Imports the pipeline function from Hugging Face’s transformers library. Pipelines are high-level wrappers for tasks like question-answering, summarization, etc.

# Load the QA model
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad") #qa_pipeline will be used to find answers to questions based on a given context.
#Initializes a question-answering pipeline.

# Extract text from PDF
#Defines a function to extract all text from a PDF file.
#pdfplumber.open(pdf_path): Opens the PDF file.
#for page in pdf.pages: Iterates through each page in the PDF.
#page.extract_text(): Extracts text content from the page.
#text += ...: Appends the extracted text to a single string, adding a newline after each page.
#return text: Returns the full text as a single string.


def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text

# Upload PDF in Google Colab
#uploaded is a dictionary where keys are filenames.
#pdf_path = list(uploaded.keys())[0]: Gets the name of the uploaded file (the first key in the dictionary).


from google.colab import files
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]

# Extract context
context = extract_text_from_pdf(pdf_path)

# Interactive question-answering loop
while True:
    question = input("Ask a question (type 'exit' to quit): ")
    if question.lower() == 'exit':
        break
    answer = qa_pipeline(question=question, context=context)['answer']
    print("Answer:", answer)
```
